```python
import re
```

```python
pattern = re.compile("\d+\s(mg|g|mcg|IU)")
re.findall(pattern , "test string g 100 mcg")
```

```python
for col in df:
    try:
        match = re.findall(pattern, df[col].iloc[7855])
        print(match)
        if len(match) > 0:
            df[col] = df[col].str.replace(f'{match[0]}', '')
            df.rename(index=str, columns = {f'{col}': f'{col}_{match[0]}'}, inplace = True)
    except:
        pass
            
```

```python
df.cholesterol = df.cholesterol.str.replace("mg", '')
df.total_fat = df.total_fat.str.replace("g", '')
df.saturated_fat = df.saturated_fat.str.replace("g", '')

df = df.rename(index = str, columns={"cholesterol": "cholesterol_mg", "total_fat": "total_fat_g", 'saturated_fat': 'saturated_fat_g'})
```

```python
df.fillna(0, inplace=True)
```

```python
for i in df:
    try:
        df[i] = df[i].astype("float")
    except:
        pass
```
